```python
import pymysql
import pandas as pd
import seaborn as sns
from jupyterthemes import jtplot
import matplotlib.pyplot as plt
```

```python
# lee el archivo JSON
df = pd.read_json('VIAJES_MARITIMOS_DIPUTADOS_ISLAS_CANARIAS.json')
df.info()
df
```

# Limpieza de datos

```python
#borramos todas las filas las cuales no tengan ninguna informacion
df=df.dropna()
df = df.drop_duplicates()
df
```

```python
#eliminacion de campos que se consideran sin relevancia para su analisis
new=df.drop(columns=['tipo','diputado_apellido2','tipo_viaje'])
new
```

```python
new = new.rename(columns={'diputado_apellido1':'apellido','diputado_nombre':'nombre'})
new
```

# Conexion y Carga a la BD

```python
#guardamos y nombramos al archivo que se creo con los datos limpios
new.to_json('archivo_limpio.json', orient='records')
new
```

```python
import pymysql
from sqlalchemy import create_engine

# detalles de conexión
engine = create_engine('mysql+pymysql://root:@localhost/my_sql')

# abre la conexión
new.to_sql('viajes', con=engine, if_exists='replace', index=False)
new
```

# Graficos

```python
datames= new.drop(columns=['anio','nombre','id_diputado','coste'])
datames= datames.groupby('apellido').count()
datames = datames.sort_values('mes',ascending=True).tail(10)
datames
```

```python
plt.title("Los 10 diputados que han realizado màs viajes en el año")
plt.pie(datames['mes'],labels= datames.index,autopct= '%0.1f %%')
plt.show
```

```python
datacoste= new.drop(columns=['anio','nombre','id_diputado','mes'])
datacoste= datacoste.groupby('apellido').sum()
datacoste = datacoste.sort_values('coste',ascending=True).tail(10)
datacoste
```

```python
plt.title("Los 10 diputados con costes anuales de viaje màs altos")
plt.pie(datacoste['coste'],labels= datacoste.index,autopct= '%0.1f %%')
plt.show
```
