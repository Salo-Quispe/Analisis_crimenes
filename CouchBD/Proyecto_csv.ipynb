# ***LIMPIEZA DE DATOS***

```python
xml = pd.read_csv('Video_Games_Sales_as_at_22_Dec_2016.csv')
xml = xml.drop(columns=['Critic_Score', 'Critic_Count','User_Score', 'User_Count','Developer','Rating'])
xml = xml.rename(columns={'Name':'Nombre',
                             'Platform':'Plataforma',
                             'Year_of_Release':'Año',
                             'Genre':'Genero',
                             'Publisher':'Editor',
                             'NA_Sales':'Ventas_NA',
                             'EU_Sales':'Ventas_EU',
                             'JP_Sales':'Ventas_JP',
                             'Other_Sales':'Otras_Ventas',
                             'Global_Sales':'Ventas_Globales'})
xml
```

```python
#Guradamos los datos limpios en otro archivo .csv 
xml.to_csv('archivo_limpiado.csv', index=False)
xml
```

```python
#en este apartado debimos covertir los archivos a jso, ya que el archivo csv
#ya que tiene valores flotantes que el los datos de couchdb no permite ya que estan fuera de rango
#para el archivo json que se crea en couchdb

import json
#tenemos que convertir los datos a json para que podamos guardar valores flotantes 
#que puedan ser guardados en couch
data = pd.read_csv('archivo_limpiado.csv')
# Convertir los datos a formato JSON
json_data = data.to_json(orient='records')

parsed_data = json.loads(json_data)
for row in parsed_data:
    for key, value in row.items():
        if isinstance(value, float):
            if abs(value) > 1.7976931348623157E+308:
                # Hacer algo para corregir el valor, como establecerlo en el valor máximo permitido en JSON
                row[key] = 1.7976931348623157E+308
                       
```

```python
#guradamos el formato a la bd            
for row in parsed_data:
    db.save(row) 
```

# ***GRAFICAS***

```python
"Grafica 1"
"En base al Editor"

editorcol = xml.drop(columns = ['Año', 'Genero', 'Nombre','Plataforma'])

editorsum = editorcol.groupby('Editor').sum()
editorsum = editorsum.sort_values(by=['Ventas_Globales'], ascending=False).head(10)
editorsum = editorsum.rename(index={'Electronic Arts':'EA',
                                    'Sony Computer Entertainment': 'Sony',
                                    'Take-Two Interactive': 'Take-Two',
                                    'Konami Digital Entertainment':'Konami',
                                    'Namco Bandai Games':'Namco'})
editorsum
```

```python
from jupyterthemes import jtplot
jtplot.style()
plt.figure(figsize=(15,8))
plt.title("Ventas por Editor")

sns.barplot(x=editorsum.index, y=editorsum['Ventas_Globales'])
```

```python
"Grafica 2"
"En base al genero"

genero = xml.drop(columns = ['Año','Nombre','Plataforma','Editor'])
genero
```

```python
generocounts = genero.groupby('Genero').count()
generocounts = generocounts.rename(index={'Role-Playing':'Roleplay'})
generocounts = generocounts.sort_values('Ventas_Globales', ascending=False)
generocounts
```

```python
jtplot.style()
plt.figure(figsize=(10,6))
plt.title("Ventas por género")

sns.barplot(x=generocounts['Ventas_Globales'], y=generocounts.index)
```

```python
"Grafica 3"
"En base a la plataforma"

plataforma = xml.drop(columns = ['Genero', 'Nombre','Año','Editor','Ventas_NA','Ventas_EU', 'Ventas_JP','Ventas_NA', 'Otras_Ventas'])

plataforma = plataforma.groupby('Plataforma').count()
plataforma = plataforma.sort_values('Ventas_Globales', ascending=False)
plataforma = plataforma.head(10)

plataforma 
```

```python
jtplot.style()
plt.figure(figsize=(12,6))
plt.title("Ventas por Plataforma")

color = sns.color_palette("hls", 8)

sns.set_palette(color)

sns.barplot(x=plataforma.index, y=plataforma['Ventas_Globales'])
```
