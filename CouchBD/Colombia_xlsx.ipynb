```python
#en este apartado debimos covertir los archivos a json, ya que el archivo xlsx
#ya que tiene valores flotantes que el los datos de couchdb no permite ya que estan fuera de rango
#para el archivo json que se crea en couchdb

import json
#tenemos que convertir los datos a json para que podamos guardar valores flotantes 
#que puedan ser guardados en couch
data = pd.read_excel('archivo_limpiado.xlsx')
# Convertir los datos a formato JSON
json_data = data.to_json(orient='records')

parsed_data = json.loads(json_data)
for row in parsed_data:
    for key, value in row.items():
        if isinstance(value, float):
            if abs(value) > 1.7976931348623157E+308:
                # Hacer algo para corregir el valor, como establecerlo en el valor máximo permitido en JSON
                row[key] = 1.7976931348623157E+308
```

```python
#guradamos el formato a la bd            
for row in parsed_data:
    db.save(row) 
```

# Graficas

```python
"Grafica 1"
"Por edad"
editorcol = data.drop(columns = ['Departamento','Clase de sitio','Arma empleada',
                               'País de nacimiento','Día','Zona'])
grafica1 = editorcol.sort_values('Edad',ascending=False).head(50)
grafica1 = grafica1.set_index('Edad')
grafica1
```

```python
plt.figure(figsize=(12,6))
plt.title("Edades afectadas por año")
color = sns.color_palette("Set2")
sns.set_palette(color)
sns.lineplot(data=grafica1, x=grafica1['Ano'], y=grafica1.index, hue='Sexo')
```

```python
"GRAFICA 2"
"Por arma empleada"
jtplot.style()
plt.title("Armas empleadas en homicidios por año")
color = sns.color_palette("Set2")
sns.set_palette(color)
sns.stripplot(data=data, x='Ano', y='Arma empleada', hue='Sexo', dodge=True)
plt.show()
```

```python
"Grafica 3"
"Por país de nacimiento"
editorcol = data.drop(columns = ['Departamento','Clase de sitio','Arma empleada','Zona','Edad','Día'])
grafica3 = editorcol.sort_values('País de nacimiento',ascending=False)
grafica3 = grafica3.set_index('País de nacimiento')
grafica3
```

```python
plt.figure(figsize=(15,30))
plt.title("Nacionalidades afectadas por año")
color = sns.color_palette("Set2")
sns.set_palette(color)
sns.violinplot(data=grafica3, x=grafica3['Ano'], y=grafica3.index)
```
